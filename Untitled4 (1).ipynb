```python
# Sales Prediction Project
# Created by: Anjali

import numpy as np
from sklearn.linear_model import LinearRegression
# sample data (TV ads budget vs sales)
X=np.array([[50],[100],[150],[200],[250]])
y=np.array([10,20,30,40,50])
#model
model=LinearRegression()
#Train
model.fit(X,y)
#Predict
prediction=model.predict([[300]])

print("Predicted Sales for 300 budget:",prediction)
print("This model predicts sales based on advertising budget.")
```
