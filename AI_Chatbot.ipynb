```python
while True:
  user = input("You:")

  if user == "hello":
    print("Chatbot: Hi!")

  elif user == "how are you":
    print("Chatbot: I'm fine!")

  elif user == "bye":
    print("Chatbot: Goodbye!")
    break

  else:
    print("Chatbot: I don't understand")
```
