```python
# smart_Recommendation_system.ipynb

import pandas as pd
data={
    "User":["A","A","A","B","B","C","C"],
    "Movie":["Inception","Titanic","Avatar","Inception","Joker","Avatar","Titanic"],
    "Rating":[5,3,4,4,5,5,2]

}
df=pd.DataFrame(data)
print(df)
```

```python
movie_matrix=df.pivot_table(index='User',columns='Movie',values='Rating')
print(movie_matrix)
```

```python
movie_similarity=movie_matrix.corr()
print(movie_similarity)
```

```python
def recommend(movie_name):
  return movie_similarity[movie_name].sort_values(ascending=False)

print(recommend("Inception"))
```

```python
movie=input("Enter a movie name:")
print(recommend(movie))
```

```python
movie=input("Enter a movie name:").title()

if movie in movie_similarity.columns:
  print(recommend(movie))
else:
  print("Movie nor found! Try another.")
```
